# 4. Filter the dataframe with only this columns

```python
df = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
df
```

# 5. Data Cleaning, a. Remove all duplicates observations

```python
df.drop_duplicates()
```

# b. Find missing values in all columns

```python
df.isna().sum()
```

# c. Remove all observations where continent column value is missing

```python
df=df[df['continent'].notna()]
```

# 6. Date time format : a. Convert date column in datetime format using pandas.to_datetime

```python
df.date = pd.to_datetime(df['date'])
df
```

# b. Create new column month after extracting month data from date column.

```python
df['month'] = pd.to_datetime(df['date']).dt.month
df
```

# 7. Data Aggregation: a. Find max value in all columns using groupby function on 'continent'column


# b. Store the result in a new dataframe named 'df_groupby'.(Use df_groupby dataframe for all further analysis)

```python
# Both a and b part are done here

df_groupby = df.groupby('continent').max()
df_groupby.reset_index()
```

# 8. Feature Engineering : a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths']/df_groupby['total_cases']
df_groupby.reset_index(inplace = True)
df_groupby
```

# 9. Data Visualization : a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

```python
import seaborn as sns
sns.distplot(df_groupby['gdp_per_capita'])
```

# b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

```python
sns.scatterplot(data = df_groupby, x = 'gdp_per_capita', y = 'total_cases')
```

# c. Plot Pairplot on df_groupby dataset.

```python
sns.pairplot(data = df_groupby)
```

# d. Plot a bar plot of 'continent' column with 'total_cases' . Tip : using kind='bar' in seaborn catplot

```python
sns.catplot(data = df_groupby,x= 'total_cases', y = 'continent',kind = 'bar')
```
